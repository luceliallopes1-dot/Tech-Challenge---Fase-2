```python
#libs - Bibliotecas importantes
#manipulação dos dados
import pandas as pd
import numpy as np

#visualização dos dados
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df_ibovespa = pd.read_csv("Ibovespa.csv")
display(df_ibovespa.head(9))
```

```python
# Converter a coluna 'Data' para datetime
df_ibovespa['Data'] = pd.to_datetime(df_ibovespa['Data'], format='%d.%m.%Y')
display(df_ibovespa.head())
```

```python
# Função para limpar e converter a coluna 'Vol.'
def clean_volume(vol):
    vol = vol.replace('.', '').replace(',', '.')
    if 'B' in vol:
        return float(vol.replace('B', '')) * 1_000_000_000
    elif 'M' in vol:
        return float(vol.replace('M', '')) * 1_000_000
    return float(vol)

df_ibovespa['Vol.'] = df_ibovespa['Vol.'].apply(clean_volume)
display(df_ibovespa.head())
```

```python
# Limpar e converter a coluna 'Var%' para numérico
df_ibovespa['Var%'] = df_ibovespa['Var%'].str.replace('%', '').str.replace(',', '.').astype(float) / 100
display(df_ibovespa.head())
```

# Task
Prever a tendência (↑ ou ↓) com acuracidade mínima de 75% em um conjunto de teste contendo o último mês (30 dias) de dados disponíveis.

## Preparação dos dados

### Subtask:
Separar os dados em conjuntos de treino e teste, onde o conjunto de teste conterá os últimos 30 dias.


**Reasoning**:
Identify the most recent date, calculate the date 30 days prior, and split the data into training and testing sets based on this date.



```python
latest_date = df_ibovespa['Data'].max()
split_date = latest_date - pd.Timedelta(days=30)

df_train = df_ibovespa[df_ibovespa['Data'] < split_date].copy()
df_test = df_ibovespa[df_ibovespa['Data'] >= split_date].copy()

print(f"Shape of training data: {df_train.shape}")
print(f"Shape of testing data: {df_test.shape}")
```

## Engenharia de recursos

### Subtask:
Criar recursos relevantes a partir dos dados históricos, como médias móveis, volatilidade e indicadores técnicos.


**Reasoning**:
Calculate the requested technical indicators (moving averages, daily change, and volatility) for both the training and testing dataframes.



```python
# Calculate features for df_train
df_train['MA_5'] = df_train['Último'].rolling(window=5).mean()
df_train['MA_20'] = df_train['Último'].rolling(window=20).mean()
df_train['Daily_Change'] = df_train['Último'] - df_train['Abertura']
df_train['Volatility_5'] = df_train['Daily_Change'].rolling(window=5).std()

# Calculate features for df_test
df_test['MA_5'] = df_test['Último'].rolling(window=5).mean()
df_test['MA_20'] = df_test['Último'].rolling(window=20).mean()
df_test['Daily_Change'] = df_test['Último'] - df_test['Abertura']
df_test['Volatility_5'] = df_test['Daily_Change'].rolling(window=5).std()

display(df_train.head(30))
display(df_test.head(30))
```

## Definição do alvo

### Subtask:
Criar a variável alvo que indica a tendência (alta ou baixa) para o dia seguinte.


**Reasoning**:
Calculate the 'Trend' column for both dataframes by comparing the current day's closing price with the next day's closing price.



```python
df_train['Trend'] = (df_train['Último'].shift(-1) > df_train['Último']).astype(int)
df_test['Trend'] = (df_test['Último'].shift(-1) > df_test['Último']).astype(int)
display(df_train.head())
display(df_test.head())
```

## Treinamento do modelo

### Subtask:
Treinar o modelo de classificação escolhido com o conjunto de treino.


**Reasoning**:
Import RandomForestClassifier, drop rows with NaN values in df_train, define features and target, instantiate RandomForestClassifier, and train the model.



```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

df_train.dropna(inplace=True)

features = df_train.select_dtypes(include=np.number).drop(columns=['Trend']).columns
target = 'Trend'

X_train = df_train[features]
y_train = df_train[target]

# model = RandomForestClassifier(random_state=42)
model = LogisticRegression(random_state=42, solver='liblinear') # Usando Regressão Logística

model.fit(X_train, y_train)
```

## Avaliação do modelo

### Subtask:
Avaliar a acuracidade do modelo no conjunto de teste.


**Reasoning**:
Evaluate the accuracy of the trained model on the test set by dropping missing values, defining features and target, making predictions, and calculating the accuracy score.



```python
from sklearn.metrics import accuracy_score

df_test.dropna(inplace=True)

X_test = df_test[features]
y_test = df_test[target]

predictions = model.predict(X_test)

accuracy = accuracy_score(y_test, predictions)

print(f"Accuracy on the test set: {accuracy:.2f}")
```
